```python
D2O50 = data16[lower:upper,:2]
D2O60 = data17[lower:upper,:2]
D2O65 = data18[lower:upper,:2]
D2O70 = data19[lower:upper,:2]
D2O80 = data20[lower:upper,:2]
D2O89 = data10[lower:upper,:2]
```

```python
def invariant(data):
    '''Calculates the invariant'''
    q = data[:,0]
    I = data[:,1]
    I[I < 0] = 0
    #invariant = integrate.simpson(q**2*I, q)
    inv = np.sqrt(np.sum(I))
    return inv
```

```python
inv50 = invariant(D2O50)
inv60 = invariant(D2O60)
inv65 = invariant(D2O65)
inv70 = invariant(D2O70)
inv80 = invariant(D2O80)
inv89 = invariant(D2O89)
invar = np.array([-inv50, -inv60, inv65, inv70, inv80, inv89])
D2O = np.array([50, 60, 65, 70, 80, 89])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
# Train the model
model.fit(D2O.reshape(-1,1), invar.reshape(-1,1))
# Make predictions
y_ = np.linspace(50, 90, 10).reshape(-1,1)
y_pred = model.predict(y_)
```

```python
plt.scatter(D2O, invar, color='k')
plt.plot(y_, y_pred)
#Fit with line to obtain PMI of silica 
plt.xlabel('Volume Fraction $D_{2}O$')
plt.ylabel('$\\sqrt{\\sum{I}}$')
plt.hlines(0, 30, 100, color = 'k', linestyles='--')
plt.xlim([50,90])
```
