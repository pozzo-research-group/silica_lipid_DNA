```python
data_Si_SANS = np.load('../Data/SANS/sub_data/np_files/r27731_Si_D2O89_1D_combined.txt.npy')[0:-50, :]
data_Si_SAXS = np.load('../Data/SANS/sub_data/np_files/Silica_SAXS.npy')[0:-50, :]
data_Si_SAXS_DOPC = np.load('../Data/SANS/sub_data/np_files/Silica_DOPC_SAXS.npy')[0:-50, :]
data_Si_DOPC = np.load('../Data/SANS/sub_data/np_files/r27730_Si_DOPC_OA00_D2O89_1D_combined.txt.npy')[0:-50, :]


sphere_model_SANS = np.load('../Data/SANS/sub_data/models/sphere_model_SANS.npy')
cs_cs_model = np.load('../Data/SANS/sub_data/models/core_shell_sphere_model.npy')


plt.rcParams.update({'font.size': 18})
#plt.errorbar(data_Si_SAXS[:,0], data_Si_SAXS[:,1], fmt='o', color='red', label='Sphere (SAXS)')


plt.errorbar(data_Si_SANS[:,0], data_Si_SANS[:,1], yerr=data_Si_SANS[:,2], fmt='o', color='C0', label='Silica')
plt.errorbar(data_Si_DOPC[:,0], data_Si_DOPC[:,1], yerr=data_Si_DOPC[:,2], fmt='o', color='C1', label='Silica with DOPC Bilayer')


plt.errorbar(sphere_model_SANS[:,0], sphere_model_SANS[:,1]*0.75, color='k', linewidth = 3, label='Sphere model')
plt.errorbar(cs_cs_model[:,0], cs_cs_model[:,1], color='green', linewidth = 3, label='Core Shell Spheres model')


#plt.errorbar(data_Si_DOPC[:,0], data_Si_DOPC[:,1], fmt='o', color='red', label='Sphere (SAXS)')

# x-axis
#plt.xlim(5e-4,1e0)
plt.xscale('log')
plt.xlabel('q ($\\AA^{-1}$)')
plt.xticks()
# y-axis
#plt.ylim(1e-3,1e6)
plt.yscale('log')
plt.ylabel('I(q) ($cm^{-1}$)')
plt.yticks()
plt.legend(fontsize=12)
#plt.savefig('../Figures/Si_DOPC_SANS_core_shell.png', dpi=600, bbox_inches="tight")
```
